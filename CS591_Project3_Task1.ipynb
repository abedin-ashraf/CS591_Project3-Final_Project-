```python
import numpy as np
import matplotlib.pyplot as plt
```

# Load the data from the provided file

```python
data = np.loadtxt('Part1_x_y_Values.txt', skiprows=1)  # Adjust file path if necessary
X = data[:, 0]
Y = data[:, 1]
```

# Normalize the data

```python
X_norm = (X - np.mean(X)) / np.std(X)  # Normalize X
Y_norm = (Y - np.mean(Y)) / np.std(Y)  # Normalize Y
```

# Gradient Descent Function

```python
def gradient_descent(X, Y, features=3, learning_rate=0.01, epochs=100):
    n = len(X)
    weights = np.random.randn(features)  # Initialize weights for the specified number of features
    weight_history = []
    loss_history = []

    for epoch in range(epochs):
        # Predictions
        Y_pred = sum(weights[i] * X**i for i in range(features))
        residuals = Y_pred - Y
        loss = np.mean(residuals**2)

        # Gradients
        grads = [(2 / n) * np.sum(residuals * X**i) for i in range(features)]

        # Update weights
        weights -= learning_rate * np.array(grads)

        # Record weights and loss
        weight_history.append(weights.copy())
        loss_history.append(loss)

        # Plot the fitting curve
        if epoch % 20 == 0 or epoch == epochs - 1:
            plt.figure(figsize=(10, 6))
            plt.scatter(X, Y, color='blue', label='Data Points')
            plt.plot(X, Y_pred, color='red', label=f'GD Curve (Epoch {epoch})')
            plt.title(f'Gradient Descent Fitting Curve (Epoch {epoch}, Features={features})')
            plt.xlabel('X')
            plt.ylabel('Y')
            plt.legend()
            plt.grid(True)
            plt.show()

    return weight_history, loss_history
```

# Stochastic Gradient Descent Function

```python
def stochastic_gradient_descent(X, Y, features=3, learning_rate=0.01, epochs=100):
    n = len(X)
    weights = np.random.randn(features)  # Initialize weights for the specified number of features
    weight_history = []
    loss_history = []

    for epoch in range(epochs):
        for i in range(n):  # Update weights for each data point
            xi, yi = X[i], Y[i]
            yi_pred = sum(weights[j] * xi**j for j in range(features))
            residual = yi_pred - yi

            # Gradients
            grads = [2 * residual * xi**j for j in range(features)]

            # Update weights
            weights -= learning_rate * np.array(grads)

        # Compute loss
        Y_pred = sum(weights[j] * X**j for j in range(features))
        loss = np.mean((Y_pred - Y)**2)

        # Record weights and loss
        weight_history.append(weights.copy())
        loss_history.append(loss)

        # Plot the fitting curve
        if epoch % 20 == 0 or epoch == epochs - 1:
            plt.figure(figsize=(10, 6))
            plt.scatter(X, Y, color='blue', label='Data Points')
            plt.plot(X, Y_pred, color='green', label=f'SGD Curve (Epoch {epoch}, Features={features})')
            plt.title(f'Stochastic Gradient Descent Fitting Curve (Epoch {epoch}, Features={features})')
            plt.xlabel('X')
            plt.ylabel('Y')
            plt.legend()
            plt.grid(True)
            plt.show()

    return weight_history, loss_history
```
